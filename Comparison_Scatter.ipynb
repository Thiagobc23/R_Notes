```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/Pokemon.csv')
df
```

```python
from matplotlib.lines import Line2D
# create a df with only legendaries
legendary = df[df['Legendary']]
# create a df with no legendaries
non = df[~df['Legendary']]

# plot non legendary
plt.scatter(non.Attack, non.Defense, color='#F85C54')
# plot legendary
plt.scatter(legendary.Attack, legendary.Defense, color='#2ACC74')

# plot legend
plt.legend(['no', 'yes'], title='Legendary')
```

```python
df['leg'] = ['Yes' if i else 'No' for i in df['Legendary']]
df['leg_color'] = ['#2ACC74' if i else '#F85C54' for i in df['Legendary']]
```

```python
plt.scatter(df.Attack, df.Defense, color=df.leg_color)

legend_elements = [Line2D([0], [0], marker='o', color='w', 
                          markerfacecolor='#F85C54', markersize=10),
                   Line2D([0], [0], marker='o', color='w', 
                          markerfacecolor='#2ACC74', markersize=10)]

plt.legend(legend_elements, ['No', 'Yes'], title='Legendary', ncol=2)
```

```python
fig, ax = plt.subplots(1, figsize=(16,8))

plt.scatter(df.Attack, df.Defense, color=df.leg_color)

legend_elements = [Line2D([0], [0], marker='o', color='w', 
                          markerfacecolor='#F85C54', markersize=10),
                   Line2D([0], [0], marker='o', color='w', 
                          markerfacecolor='#2ACC74', markersize=10)]

legend = plt.legend(legend_elements, ['No', 'Yes'], 
           title='Legendary', ncol=2, loc='lower left',
           bbox_to_anchor=(0.45, -0.15), frameon=False, fontsize=12)

plt.setp(legend.get_title(),fontsize=14)
legend.get_title().set_position((-110, -18))
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize=(16,8), facecolor="#424242")
ax.set_facecolor("#424242")

plt.scatter(df.Attack, df.Defense, color=df.leg_color, s=15)

legend_elements = [Line2D([0], [0], marker='o', color="#424242", 
                          markerfacecolor='#F85C54', markersize=10),
                   Line2D([0], [0], marker='o', color="#424242", 
                          markerfacecolor='#2ACC74', markersize=10)]

legend = plt.legend(legend_elements, ['No', 'Yes'], 
                    title='Legendary', ncol=2, 
                    loc='lower left', frameon=False,
                    bbox_to_anchor=(0.45, -0.15), 
                    fontsize=12)

legend.get_texts()[0].set_color('w')
legend.get_texts()[1].set_color('w')

plt.setp(legend.get_title(),fontsize=14, color='w')
legend.get_title().set_position((-110, -18))

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.spines['left'].set_color('w')
ax.spines['bottom'].set_color('w')

for tick in ax.xaxis.get_ticklines(): 
    tick.set_color('w')
for tick in ax.yaxis.get_ticklines(): 
    tick.set_color('w')
    
plt.xticks(color='w')
plt.yticks(color='w')

plt.title('Stats values from Pokemon Games (Not Pokemon Cards or Pokemon GO\n', 
          loc='left', color='w')
plt.suptitle('Pokemon Attack X Defense', x=0.125, ha='left', 
             fontsize=16, color='w')
plt.text(200,-50,'Source: https://www.kaggle.com/abcsds/pokemon', ha='right', fontsize=11, color='w')

plt.xlabel('Attack', color='w', fontsize=12)
plt.ylabel('Defense', color='w', fontsize=12)

plt.hlines(df.Defense.mean(), -5, 200, color='w', alpha=0.5)
plt.ylim(-5,250)
plt.vlines(df.Attack.mean(), -5, 300, color='w', alpha=0.5)
plt.xlim(-5,200)

arrowprops = dict(arrowstyle="->", connectionstyle="angle3,angleA=0,angleB=-90", color='w')
plt.annotate('Shuckle', 
             xy=(10, 230), 
             xytext=(10+3, 230-15), 
             arrowprops=arrowprops,
             fontsize = 12, color='w')

plt.annotate('Deoxys', 
             xy=(180, 20), 
             xytext=(180-15, 10+25), 
             arrowprops=arrowprops,
             fontsize = 12, color='w')

# grid
ax.set_axisbelow(True)
ax.yaxis.grid(color='#848484', linestyle='solid')
ax.xaxis.grid(color='#848484', linestyle='solid')

plt.show()
```
